```python
import os
import numpy as np
import pandas as pd
```

```python
encuesta = pd.read_csv('Dataframe_unido.csv', index_col = 'index')
encuesta.head(2)
```

```python
# Renombramos algunas columnas importantes con las que vamos a trabajar
encuesta.rename(columns = {'Q3': 'paises', 'Q4': 'título_académico', 'Q5': 'puesto_trabajo', 'Q6': 'años_experiencia', 'Q8': 'lenguaje_programación', 'Q11': 'herramientas_empresas', 'Q25': 'salario'}, inplace = True)
```

```python
encuesta.head()
```

```python
# Eliminamos la columna time porque no nos interesa
encuesta = pd.DataFrame(encuesta.drop('time', axis = 1))
encuesta.head()
```
